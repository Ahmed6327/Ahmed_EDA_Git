# $5.$ $Outlier$ $Analysis$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path="train_ctrUa4K.csv"
train_df = pd.read_csv(path)
train_df
```

**Steps in Outlier analysis**

- Step-1: Find the Q1 , Q2 and Q3
    
  - np.percentile(column data, q)

- Step-2: Calculate the IQR
    
  - IQR= Q3-Q1

- Step-3: Calculate lower boundary and upper boundary
    
  - lb: Q1-1.5IQR
    
  - ub: Q3+1.5IQR
    
- Step-4: Find the Outliersdf
    
  - c1: column data < lb
    
  - c2: column data > ub
    
  - c: apply the main condition
    
  - main data[c]

```python
#####------------- Step-1------###########


Q1=np.percentile(train_df['ApplicantIncome'],25)
Q2=np.percentile(train_df['ApplicantIncome'],50)
Q3=np.percentile(train_df['ApplicantIncome'],75)

#####------------- Step-2------###########

IQR=Q3-Q1

#####------------- Step-3------###########

lb=Q1-1.5*IQR
ub=Q3+1.5*IQR

#####------------- Step-4------###########

c1=train_df['ApplicantIncome']<lb
c2=train_df['ApplicantIncome']>ub
con=c1|c2

#####------------- Step-5------###########

outliers_df=train_df[con]
outliers_df

#####------------- Step-6------###########

c1=train_df['ApplicantIncome']>lb
c2=train_df['ApplicantIncome']<ub
con=c1&c2
non_outliers_df=train_df[c1&c2]
non_outliers_df
```

**Compare original data with non outliers data**

will plot histogram and box plot of the both

```python
plt.hist(train_df['ApplicantIncome'])
plt.hist(non_outliers_df['ApplicantIncome'])
plt.show()
```

```python
plt.boxplot(train_df['ApplicantIncome'])
plt.boxplot(non_outliers_df['ApplicantIncome'])
plt.show()
```

```python
#  method_1

plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.hist(train_df['ApplicantIncome'],bins=40)
plt.subplot(2,2,2)
plt.boxplot(train_df['ApplicantIncome'])
plt.subplot(2,2,3)
plt.hist(non_outliers_df['ApplicantIncome'],bins=40)
plt.subplot(2,2,4)
plt.boxplot(non_outliers_df['ApplicantIncome'])
plt.show()
```

## **How to deal outliers**

**Drop the outliers**

- we can drop the outliers if outlier percentage < 2%

- But this is not recommended , we lost other columns data also

**Impute with Median values**

- As we know that Median does not affect by outliers

- So it is good prcatice we can impute outliers with Median value

**Cap with Q3 or Q1 value**

- If outliers are presnt less than lower bound then fill with Q1

- If outliers are more than upper bound then fill with Q3

```python
#####------------- Step-1------###########

Q1=np.percentile(train_df['ApplicantIncome'],25)
Q2=np.percentile(train_df['ApplicantIncome'],50)
Q3=np.percentile(train_df['ApplicantIncome'],75)

#####------------- Step-2------###########

IQR=Q3-Q1

#####------------- Step-3------###########

lb=Q1-1.5*IQR
ub=Q3+1.5*IQR

median=train_df['ApplicantIncome'].median()

list1=[]
for i in train_df['ApplicantIncome']:
    if i<lb or i>ub:
        list1.append(median)
    else:
        list1.append(i)
        
train_df['ApplicantIncome_new']=list1
```

```python
plt.boxplot(train_df['ApplicantIncome_new'])
plt.show()
```

**np.where**

```python
path="train_ctrUa4K.csv"
train_df = pd.read_csv(path)
train_df
```

- above replace one we use a traditional approach

- for loop, list, if-else

- the same we can get by using np.where method

```python
#####------------- Step-1------###########

Q1=np.percentile(train_df['ApplicantIncome'],25)
Q2=np.percentile(train_df['ApplicantIncome'],50)
Q3=np.percentile(train_df['ApplicantIncome'],75)

#####------------- Step-2------###########

IQR=Q3-Q1

#####------------- Step-3------###########

lb=Q1-1.5*IQR
ub=Q3+1.5*IQR

median=train_df['ApplicantIncome'].median()

c1=train_df['ApplicantIncome']<lb
c2=train_df['ApplicantIncome']>ub
con=c1|c2

t=median
f=train_df['ApplicantIncome']

train_df['ApplicantIncome']=np.where(con,t,f)
plt.boxplot(train_df['ApplicantIncome'])
plt.show()
```
